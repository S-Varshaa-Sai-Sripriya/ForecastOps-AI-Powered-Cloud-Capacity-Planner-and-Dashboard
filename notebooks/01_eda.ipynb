```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv("../data/capacity_usage_simulated.csv", parse_dates=["timestamp"])
```

```python
print("Regions available:", df["region"].unique())
```

```python
region = "us-east"
df_region = df[df["region"] == region]
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(data=df_region, x="timestamp", y="cpu_usage", label="CPU Usage")
sns.lineplot(data=df_region, x="timestamp", y="memory_usage", label="Memory Usage")
plt.title(f"CPU vs Memory Usage in {region}")
plt.xlabel("Time")
plt.ylabel("Usage (%)")
plt.legend()
plt.tight_layout()
plt.show()
```
